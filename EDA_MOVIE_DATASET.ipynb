### top voted movie

```python
def topvoted_movie(x):
    votes=[]
    votes=data.sort_values(by=['votes'], ascending=False).head(x)['votes']
    top_voted_movies=[]
    top_voted_movies=data.sort_values(by=['votes'], ascending=False).head(x)['MOVIES']
    d1=zip(top_voted_movies,votes)
    return dict(d1)
a=topvoted_movie(5)
a
```

### year

```python
data["Year"] = data["YEAR"].str.replace('(', '')
data["Year"] = data["YEAR"].str.replace(')', '')
```

```python
data['Year'] = data['Year'].str.strip()
data[["Mix", "End_Year"]] = data['Year'].str.split('–', 1, expand=True)
```

```python

import re

data['Start_Year'] = data['Mix'].astype(str).apply(lambda x: ''.join(re.findall(r'\d{4}', x)))
data['Season'] = data['Mix'].astype(str).apply(lambda x: ''.join(re.findall(r'\b[I|V|X|L|C|D|M]+\b', x)))
data['Type'] = data['Mix'].astype(str).apply(lambda x: ''.join(re.sub(r'(\d{4}|\b[I|V|X|L|C|D|M]+\b)', '', x)))
```
